```python
tweaks = ["primary=28,residential=18,secondary=23,tertiary=19,trunk=43,unclassified=12,","primary=21,residential=21,secondary=15,tertiary=23,trunk=19,unclassified=10,","primary=30,secondary=26,","primary=25,"]
```

```python
tweaks[0]
```

```python
previous = 0
current = 1

for each in tweaks:
    # creating a variable and storing the text
    # that we want to search
    search_text = tweaks[previous]

    # creating a variable and storing the text
    # that we want to add
    replace_text = tweaks[current]

    # Opening our text file in read only
    # mode using the open() function
    with open(r'/home/lohitd@nextbillion.ai/Downloads/osrm-backend-master/profiles/car.lua', 'r') as file:

        # Reading the content of the file
        # using the read() function and storing
        # them in a new variable
        data = file.read()

        # Searching and replacing the text
        # using the replace() function
        data = data.replace(search_text, replace_text)

    # Opening our text file in write only
    # mode to write the replaced content
    with open(r'/home/lohitd@nextbillion.ai/Downloads/osrm-backend-master/profiles/car.lua', 'w') as file:

        # Writing the replaced data in our
        # text file
        file.write(data)
  
# Printing Text replaced
print("Text replaced")
```
